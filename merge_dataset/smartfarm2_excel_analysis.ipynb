```python
import numpy as np
import pandas as pd
```

```python
# KOR to ENG
kor_dict = {
    "품명": "Product Name",
    "번호": "Number",
    "과장": "Fruit vertical length",
    "과폭": "Fruit horizontal length",
    "과중": "Fruit weight",
    "경도": "Hardness",
    "당도": "Sweetness",
    "산도": "Acidity",
    "죽향 (실온)": "Jukhyang  (Room Temperature)",
    "죽향 (저장)": "Jukhyang  (Storage)",
    "메리퀸": "Maryqueen",
    "색 착 계": "Color",
    "비고": "Remarks"
}
```

```python
# Section translated to number
section_dict = {
    "①": 1,
    "②": 2,
    "③": 3,
    "④": 4,
    "⑤": 5,
    "⑥": 6
}
```

```python
excel_file = 'fruitSM2.xlsx'
```

```python
xls = pd.ExcelFile(excel_file)
```

```python
# list all sheet names
xls.sheet_names
```

```python
# get only sheets with format: '24.10.14' as year.month.day, then sort them
sheets = [sheet for sheet in xls.sheet_names if sheet[0].isdigit()]
sheets.sort()
sheets
```

```python
# read the sheet 
df = pd.read_excel(xls, sheets[0])
print(sheets[0])    
```

```python
date = df.columns[1].split()[-1]
date
```

```python
df.head()
```

```python
cols = df.iloc[0].tolist()[1:]
print(cols)
# translate to English
cols = [kor_dict.get(col, col) for col in cols]
en_name = '색 착 계'
cols[-4:-1] = [kor_dict[en_name] + ' L', kor_dict[en_name] + ' a', kor_dict[en_name] + ' b']
print(cols)
cols = ['Product Name', 'Section', 'Plant', 'Fruit vertical length', 'Fruit horizontal length', 'Fruit weight', 'Hardness', 'Sweetness', 'Acidity', 'Color L', 'Color a', 'Color b', 'Remarks']
```

```python
# new dataframe 
new_df = pd.DataFrame(columns=cols)
new_df
```

```python
product_name = np.nan
for i in range(2, len(df)):
    row = df.iloc[i].tolist()
    if pd.isnull(row[1]):
        row[0] = product_name
    else:
        # remove the /n in the product name
        row[1] = row[1].replace('\n', '')
        product_name = row[1]
    number = row[2].split('-')
    section = section_dict[number[0]]
    plant = int(number[1])
    new_row = [product_name, section, plant] + row[3:]
    new_df.loc[i-2] = new_row
new_df
```

```python
report_df = {}
for i in sheets:
    df = pd.read_excel(xls, i)
    date = df.columns[1].split()[-1]
    print(date)
    cols = ['Product Name', 'Section', 'Plant', 'Fruit vertical length', 'Fruit horizontal length', 'Fruit weight', 'Hardness', 'Sweetness', 'Acidity', 'Color L', 'Color a', 'Color b', 'Remarks']
    new_df = pd.DataFrame(columns=cols)
    product_name = np.nan
    for i in range(2, len(df)):
        row = df.iloc[i].tolist()
        if pd.isnull(row[1]):
            row[0] = product_name
        else:
            row[1] = row[1].replace('\n', '')
            product_name = row[1]
        number = row[2].split('-')
        section = section_dict[number[0]]
        plant = int(number[1])
        new_row = [product_name, section, plant] + row[3:]
        new_df.loc[i-2] = new_row
    report_df[date] = new_df
    
```

```python
# export to csv files to path
path = 'fruitSM2/'
for date, df in report_df.items():
    df.to_csv(path + date + '.csv', index=False, encoding='utf-8-sig')    
```
